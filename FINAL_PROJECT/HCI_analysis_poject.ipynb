**Описание признаков:**

Индекс человеческого капитала содержит в себе влияние факторов:

**Probability of Survival to Age 5** - Вероятность дожития до 5 лет рассчитывается путем вычитания коэффициента смертности детей в возрасте до 5 лет из 1. 

**Expected Years of School** - Ожидаемые годы обучения в школе рассчитываются как сумма показателей зачисления в зависимости от возраста в возрасте от 4 до 17 лет.

**Harmonized Test Scores** - Согласованные результаты тестов  (школьных экзаменов), объединяет баллы по основным международным программам тестирования успеваемости учащихся, измеряемые в единицах, эквивалентных TIMMS, где 300 - минимальное достижение, а 625 - продвинутое достижение. 

**Learning-Adjusted Years of School**** - Школьные годы с поправкой на обучение рассчитываются путем умножения оценок ожидаемых лет обучения на отношение самых последних согласованных результатов тестов к 625, где 625 соответствует продвинутому уровню по тесту TIMSS (Тенденции в международном изучении математики и естественных наук), основанному на методологии Filmer et al. (2018 год).

**Fraction of Children Under 5 Not Stunted** - Доля детей в возрасте до 5 лет, не страдающих задержкой роста, рассчитывается путем вычитания показателей задержки роста из 1. 

**Adult Survival Rate** - Коэффициент выживаемости взрослого населения рассчитывается путем вычитания коэффициента смертности для 15-60-летних из 1. 

**HUMAN CAPITAL INDEX 2020 (HCI)** - рассчитывает вклад здравоохранения и образования в производительность труда работников. Итоговый показатель индекса колеблется от нуля до единицы и измеряет производительность труда будущего работника ребенка, родившегося сегодня, по сравнению с показателем полного здоровья и полного образования.

**HUMAN CAPITAL INDEX 2020 (LOWER BOUND)** - Нижняя граница HCI отражает неопределенность в измерении компонентов и общего индекса. Он получается путем пересчета HCI с использованием оценок нижних границ каждого из компонентов HCI. Диапазон между верхней и нижней границами - это интервал неопределенности. Хотя построенные здесь интервалы неопределенности не имеют строгой статистической интерпретации, эмпирическое правило состоит в том, что если для двух стран они существенно перекрываются, различия между их значениями HCI вряд ли будут практически значимыми.

**HUMAN CAPITAL INDEX 2020 (UPPER BOUND)** - Верхняя граница HCI отражает неопределенность в измерении компонентов и общего индекса. Он получается путем пересчета HCI с использованием оценок верхних границ каждого из компонентов HCI. Диапазон между верхней и нижней границами - это интервал неопределенности. Хотя построенные здесь интервалы неопределенности не имеют строгой статистической интерпретации, эмпирическое правило состоит в том, что если для двух стран они существенно перекрываются, различия между их значениями HCI вряд ли будут практически значимыми."

```python
# Странные значения!
# Признак 'Fraction of Children Under 5 Not Stunted' - это количественный показатель, который выражен числом. 
# Но в описании датасета почему-то указа тип object (!?)
# Попытка преобразовать весь столбец к типу float дала ошибку. Значит, в поле встречаются нечисловые значения.
# Отсортирую значения столбца и список уникальных значений, чтоб увидеть, какие значения не являются числами
df['Fraction of Children Under 5 Not Stunted'].sort_values().unique()
```

```python
# Сортировка значений показала, что среди значений в столбце есть прочерки==>
# ==> нужно понять , насколько много прочерков и можно ли их заменить на числовые значения ==>
# Проведу анализ значений столбца с промощью функции

def dataset_tranformation(cols, ctg):
# ====================Функция: ========================
# -  убирает из столбца все 'нечисла' 
# - возвращает список 'нечисел' 
# - сoздает датафрейм  без строк , которые содежат 'нечисла'
# - преобразует проблемный столбец к типу float 
# - выводит статистику по числовому признаку в разрезе категорий 
# cols - признак, который надо заменить
# ctg - категориальный признак, по которому группируется статистика по количественному признаку
    filter_data = df
    strange_simbols_list = []
    singl_col_list = df[cols].unique().tolist()
    for item in singl_col_list:
        try:
            float(item)
        except:
            TypeError
            strange_simbols_list.append(item)
            print(f"Странное значение <{item}> содержится в {df[cols][df[cols] == item].count()} стр.")
            filter_data = df.drop(np.where(df[cols] == item)[0])
    filter_data = filter_data.astype({cols: np.float64})
    print(f"Медианные значение признака {cols}:")
    filter_data = pd.pivot_table(filter_data, index=[ctg], values=[cols], aggfunc=np.median)
    filter_data = pd.concat([filter_data, 
                             pd.DataFrame([filter_data.median(axis=0)], index=['All'])])
    filter_data = filter_data.sort_values(by=cols)
    filter_data = filter_data.sort_values(by=cols)
    filter_data = filter_data.rename_axis(ctg).reset_index()
    return filter_data
    
f_pivot = dataset_tranformation('Fraction of Children Under 5 Not Stunted', 'Region')       
f_pivot
```

Вывод: 
- можно отбросить весь столбец 'Fraction of Children Under 5 Not Stunted', так как примерно половина значений - это пропуски.  
- можно заменить пропуски медианным признаком в зависимости от региона и оценить корреляцию с целевым признаков. По результатам приянть решение об использовании данного признака в модели

Я решила отработать по второму варианту.

```python
# замена пропусков 'Fraction of Children Under 5 Not Stunted' на медианное значение
def change_data(cols, ctg, feach_list, stat_frame):
# =================Функция заменяет значения в любом столбце по заданному условию ===========================
# cols - признак, который надо заменить
# ctg - категориальный признак, по которому производится поиск значения для замены
# feach_list - список всех значений категориального признака
# stat_frame - статитический датафрей, полученный с помощью функции dataset_tranformation
    ctg_list = stat_frame[ctg].tolist()
    df_changed = df
    for i in feach_list:
        if cols == ctg:
            df_changed[cols].where(~(df_changed[cols]==i), other=str(ctg_list.index(i)+1), inplace=True)
        else:
            if i in ctg_list:
                df_changed[cols].where(~((df_changed[ctg]==i) & (df_changed[cols]=='-')), other=str(stat_frame.iloc[ctg_list.index(i), 1]), inplace=True)
            else:
                df_changed[cols].where(~((df_changed[ctg]==i) & (df_changed[cols]=='-')),
                                                         other=str(stat_frame.iloc[ctg_list.index('All'), 1]), inplace=True)
    
    #преобразование признака к типу float
    df_changed = df_changed.astype({cols: np.float64})
    return df_changed
df = change_data('Fraction of Children Under 5 Not Stunted', 'Region', regions, f_pivot)
df.info()         
```

## 4. Обучение моделей, проверка качества модели по R2

```python
scorring = []
def model_score_and_visual(feach_test, true_values, pred_values):
#Расчет метрики R2 и визуализация модели
    r2_metric = round(r2_score(true_values, pred_values), 4)
    print(f"\033[35m\033[1m R2: {r2_metric} \n")
    
    f_tsne = tsne_model.fit_transform(feach_test)
    f_tsne = f_tsne.flatten().tolist()   
    
    plt.figure(figsize=(5, 5))
    
    plt.scatter(x=f_tsne, y=true_values.tolist(), label='true values')
    plt.scatter(x=f_tsne, y=pred_values, color='r', label='predicted values')
    plt.ylim(0, 1)
    plt.legend()
    plt.xlabel('Features(tsne_agrigeted)')
    plt.ylabel('Target')
    plt.title('True vs Predicted values')
       
    plt.show()
    return r2_metric
```

```python
#линейная регрессия (LinearRegression)
linr_model = LinearRegression()
linr_model.fit(X_train, y_train)
y_pred = linr_model.predict(X_test)
scorring.append(model_score_and_visual(X_test, y_test, y_pred))
```

```python
# проработка гиперпараметров для алгоритма "случайный лес" (RandomForestRegressor)
rfr_parameters = [{'n_estimators': [5, 10], 'max_features': np.arange(3, 7), 'max_depth': np.arange(3, 7)}]
rfr_model = RandomForestRegressor()
rfr_model_param_selection = GridSearchCV(estimator=rfr_model, param_grid=rfr_parameters, scoring='r2', cv=3)
rfr_model_param_selection.fit(X_train, y_train)
rfr_model_param_selection.best_params_
```

```python
# случайный лес (RandomForestRegressor)
rfr_model = RandomForestRegressor(n_estimators=5, max_features=3, max_depth=6, random_state=42, n_jobs=3)
rfr_model.fit(X_train, y_train)
y_pred = rfr_model.predict(X_test)
scorring.append(model_score_and_visual(X_test, y_test, y_pred))
```

```python
# градиентный бустинг (GradientBoostingRegressor)
gbr_model = GradientBoostingRegressor(n_estimators=5, max_features=3, max_depth=6, criterion='mse', random_state=42)
gbr_model.fit(X_train, y_train)
y_pred = gbr_model.predict(X_test)
scorring.append(model_score_and_visual(X_test, y_test, y_pred))
```
