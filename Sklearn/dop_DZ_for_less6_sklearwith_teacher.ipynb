```python
X = pd.DataFrame(data['data'], columns = data['feature_names'])
X.sample()
```

```python
X.info()
```

Датафрейм содержит 178 строк и 13 столбцов со значениями признаков
Пропущенных значений нет

```python
X['target'] = data['target']
X.sample()
```

```python
X_corr = X.corr()
X_corr
```

```python
high_corr = list(X_corr[(X_corr['target']<1) & (X_corr['target']>0.5)].index)
high_corr
```

```python
X = X.drop(['target'], axis=1)
X.sample()
```

```python
high_corr_new = list(i+'_2' for i in high_corr)
high_corr_new
```

```python
X[high_corr_new] = X[high_corr]**2
X.sample()
```
