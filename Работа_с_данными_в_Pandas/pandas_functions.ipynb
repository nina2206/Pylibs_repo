# Задание 1

```python
import pandas as pd
import numpy as np
authors = pd.DataFrame({'author_id': [1, 2, 3], 
                        'author_name': ['Тургенев', 'Чехов', 'Островский']})
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                    'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 
                    'price': [450, 300, 350, 500, 450, 370, 290]})
books
```

# Задание 2

```python
authors_price = pd.merge(authors, books, on = 'author_id')
authors_price
```

# Задание 3

```python
top5 = authors_price.sort_values(by="price", ascending=False).head(5)
top5
```

# Задание 4

```python
authors_stat = authors_price.pivot_table(index="author_name", values=["price"], 
                                         aggfunc=[np.min, np.max, np.mean])
authors_stat = authors_stat.rename(columns = {'amin': 'min_', 'amax': 'max_', 'mean': 'mean_'})
authors_stat            
```
